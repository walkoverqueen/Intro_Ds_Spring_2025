```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#load dataset
df = pd.read_csv("Cleaned_Chicago_Schools_Data.csv")
df
```

#### Data Visualization: Scatter Plot
**Teacher Score vs Average Student Attendance** <br>
Explores if schools with higher teacher quality have better student attendance.


```python
plt.figure(figsize=(7, 5))
plt.scatter(df['Teachers Score'], df['Average Student Attendance'], color='blue')
plt.title("Teacher Score vs Student Attendance")
plt.xlabel("Teachers Score")
plt.ylabel("Average Student Attendance (%)")
plt.grid(True)
plt.show()

plt.savefig('teachscore_v_attendance.jpg')
```

#### Data Visualization: Pie Chart
**Proportion of Chicago Schools Meeting Adequate Yearly Progress**

```python
progress_counts = df['Adequate Yearly Progress Made?'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(progress_counts, labels=progress_counts.index, autopct='%1.1f%%', startangle=90)

#title
plt.title("Proportion of Schools Meeting Adequate Yearly Progress")

plt.ylabel("")  # or use plt.ylabel("Your Custom Label") if you still want something there
plt.show()

plt.savefig('adequate_yearly_progress.jpg')
```
