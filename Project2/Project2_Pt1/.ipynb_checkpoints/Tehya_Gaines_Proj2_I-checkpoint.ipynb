##### 3.3: Use groupby() to compute summary statistics
These examples compare categories by aggregating numerical data

```python
#avg safety score of elementary, middle, and high schools
df.groupby("Elementary, Middle, or High School")["Safety Score"].mean()
```

```python
#counts how many schools did or didnt make yearly progress goals
df.groupby("Adequate Yearly Progress Made? ")["Name of School"].count()
```

```python
#avg rate of attendance rate for schools that made adequate yearly progress and thos that didnt.
df.groupby("Adequate Yearly Progress Made? ")["Average Student Attendance"].mean()
```
